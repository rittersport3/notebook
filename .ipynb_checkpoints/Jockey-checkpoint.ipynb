```python
ENDERECO_JCK = "http://www.jockeysp.com.br/"
HIP = u'Hip\xf3dromo:\r\n                  C.JARDIM\r\n                  '
```

```python
def retorna_paginas():
    """ Encontra, para todos os ranges, todos os campos a """
    for a in range(2012, 2015):
        sopa = bs4.BeautifulSoup(qry_data(1,a).content, from_encoding=qry_data(1,a).encoding)
        yield sopa.findAll("a")
```

```python
def retorna_link(links):
    """ Função que retorna todos os links dentro dos links encontrados """
    for i in links: 
        yield ENDERECO_JCK + i['href']
```
